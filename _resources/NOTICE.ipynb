## Recolección de datos
Para mejorar la experiencia de los usuarios y la calidad de los activos de dbdemos, dbdemos envía reportes de uso y captura vistas en el notebook instalado (usualmente en la primera celda) y otros activos como dashboards. Esta información se recopila solo para la mejora del producto y no con fines de marketing, y no contiene información personal identificable (PII). Al usar `dbdemos` y los activos que provee, consientes esta recolección de datos. Si deseas desactivarla, puedes establecer `Tracker.enable_tracker` en False en el archivo `tracker.py`.

## Creación de recursos
Para simplificar tu experiencia, `dbdemos` creará y pondrá en marcha recursos por ti. Por ejemplo, una demo podría iniciar (no exhaustivo):
- Un clúster para ejecutar tu demo
- Una canalización Delta Live Table para ingerir datos
- Un endpoint DBSQL para ejecutar dashboards DBSQL
- Un modelo de ML

Aunque `dbdemos` hace su mejor esfuerzo para limitar el consumo y forzar la auto-terminación de recursos, tú eres responsable de los recursos creados y del posible consumo asociado.

## Soporte
Databricks no ofrece soporte oficial para `dbdemos` ni para los activos asociados.
Para cualquier problema con `dbdemos` o las demos instaladas, por favor abre un issue y el equipo de demos lo revisará en la medida de lo posible.
